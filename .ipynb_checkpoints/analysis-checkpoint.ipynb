```python
# Рисовать графики сразу же
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("budgetdata.csv",sep=";",encoding="cp1251")
pd.unique(df["budget"])
df[df["budget"]==153]
df.loc[df["budget"]==153, ['kesr', 'summa']]
df.groupby("kvsr")["summa"].sum().plot(kind='bar', fontsize=10,figsize=(8, 8),width=1)
```

```python
df.groupby("budget")["summa"].sum().plot(kind='pie', fontsize=8,figsize=(8, 8))
```
